```python
#How do I remove columns from a pandas DataFrame
import pandas as pd
```

```python
# read a dataset of UFO reports into a DataFrame
ufo=pd.read_csv('ufo.csv')
```

```python
ufo
```

```python
#Remove a single column(axis=1, refer to columns)
ufo.drop('Colors Reported',inplace=True,axis=1)
```

```python
ufo
```

```python
#Remove multiple column at once
ufo.drop(['Time','State'],inplace=True,axis=1)
```

```python
ufo
```

```python
# remove multiple rows at once (axis=0 refers to rows)
ufo.drop([0,3,5],inplace=True,axis=0)
ufo
```
